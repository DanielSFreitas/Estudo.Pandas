```python
import pandas as pd
bank = pd.read_csv('/content/drive/MyDrive/Atividade/bank.csv', sep = ';')
bank.head(10)
```

```python
bank.groupby(['marital', 'job']).mean()
import numpy as np
bank.pivot_table(index = 'marital', columns= ['job'], aggfunc=np.mean)
```

```python
divorced_data = bank[bank['marital'] == 'divorced']
bank['percentual_entrepreneurs']= (divorced_data['job'] =='entrepreneur').mean()*100
percentual_entrepreneurs = (divorced_data['job'] =='entrepreneur').mean()*100
percentual_students = (divorced_data['job'] == 'student').mean()*100
```

```python
print(f"Percentual de divorciados que são empreendedores: {percentual_entrepreneurs:.2f}%")
print(f"Percentual de divorciados que são estudantes: {percentual_students:.2f}%")
```

```python
bank['tops'] =pd.qcut(bank['balance'], q=[0 ,0.8, 0.9, 1], labels=['80% Inferiores','Gold', 'Premium'])
```

```python
bank['premium_count'] = (bank['tops'] == 'Premium').sum()
gold_count = (bank['tops'] == 'Gold').sum()
```

```python
top_10_clientes = bank.sort_values(by='balance', ascending=False).head(10)
colunas_relevantes = ['age', 'job', 'marital', 'education', 'default', 'loan', 'duration']
tabela_final = top_10_clientes[colunas_relevantes]

def highlight_max_min(data):
    styles = data.copy()
    for column in data.columns:
        if column == 'duration':
            max_val = data[column].max()
            min_val = data[column].min()
            styles[column] = [
                'background-color: red' if v == max_val else
                'background-color: yellow' if v == min_val else
                '' for v in data[column]
            ]
        else:
            styles[column] = ''
    return styles

styled_df = tabela_final.style.apply(highlight_max_min, axis=None)
styled_df
```
