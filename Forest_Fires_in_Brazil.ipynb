```python
#Calculando o número de incêncios no Mato Grosso no ano de 2010
Mato_grosso_incendio = df[((df['year']==2010) & (df['number']>0) & df['state'].str.contains('Mato Grosso'))]
Mato_grosso_incendio['number'].sum()
```

```python
#Calculando a média e o desvio padrão de incêncios por mês no Acre
Acre= df[((df['number']>0) & df['state'].str.contains('Acre'))]
Acre[['month','number']].mean()
Acre[['month','number']].std()
```

```python
# Calculando a média dos números de incêndios e meses para cada estado
df.groupby('state')['number', 'month'].mean()
```

```python
# Calculando a média dos números e meses para cada estado
media_por_estado = df.groupby('state')[['number', 'month']].mean()

# Encontrando o estado com a maior média de incêndios por mês
estado_maior_media = media_por_estado['number'].idxmax()
maior_media = media_por_estado['number'].max()

# Encontrando o estado com a menor média de incêndios por mês
estado_menor_media = media_por_estado['number'].idxmin()
menor_media = media_por_estado['number'].min()

print(f"Estado com a maior média: {estado_maior_media} ({maior_media} incêndios por mês)")
print(f"Estado com a menor média: {estado_menor_media} ({menor_media} incêndios por mês)")
```

```python
#Criação de coluna sudeste e centro oeste
sudeste = df['state'].isin(['Espirito Santo', 'Minas Gerais', 'Rio', 'Sao Paulo'])
centro_oeste = df['state'].isin(['Goias', 'Mato Grosso', 'Mato Grosso', 'Distrito Federal'])

import numpy as np
df['regiao_sudeste'] = np.where(sudeste, 1, 0)
df['regiao_centro_oeste'] = np.where(centro_oeste, 1, 0)
```

```python
#Numero de incendios na região sudeste
incendio_sudeste = df[((df['regiao_sudeste']==1) & (df['number']>0))]
incendio_sudeste['number'].sum()
```

```python
#Numero de incendios na região centro_oeste
incendio_centro_oeste = df[((df['regiao_centro_oeste']==1) & (df['number']>0))]
incendio_centro_oeste['number'].sum()
```
