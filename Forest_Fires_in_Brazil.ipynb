```python
# Calculando a mÃ©dia dos nÃºmeros e meses para cada estado

df.groupby(['state', 'month'])['number'].mean()
```

```python
# Calculando a mÃ©dia dos nÃºmeros e meses para cada estado
df.groupby('state')['number', 'month'].mean()
```

```python
# Calculando a mÃ©dia dos nÃºmeros e meses para cada estado
media_por_estado = df.groupby('state')[['number', 'month']].mean()

# Encontrando o estado com a maior mÃ©dia de incÃªndios por mÃªs
estado_maior_media = media_por_estado['number'].idxmax()
maior_media = media_por_estado['number'].max()

# Encontrando o estado com a menor mÃ©dia de incÃªndios por mÃªs
estado_menor_media = media_por_estado['number'].idxmin()
menor_media = media_por_estado['number'].min()

print(f"Estado com a maior mÃ©dia: {estado_maior_media} ({maior_media} incÃªndios por mÃªs)")
print(f"Estado com a menor mÃ©dia: {estado_menor_media} ({menor_media} incÃªndios por mÃªs)")
```
