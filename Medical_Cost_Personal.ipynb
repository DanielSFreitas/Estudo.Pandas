```python
def categorizar_bmi_male(bmi):
    if bmi < 20:
        return 'abaixo do normal'
    elif 20 <= bmi <= 24.9:
        return 'normal'
    else:
        return 'acima do normal'

df['bmi_male'] = df['bmi'].apply(categorizar_bmi_male)
```

```python
homens_bmi_alto = df[(df["bmi_male"]=="acima do normal") & (df['sex'] == 'male')].value_counts()
qtd = len(homens_bmi_alto)
qtd
```
