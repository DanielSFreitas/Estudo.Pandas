```python
import pandas as pd
prod = pd.read_csv('/content/drive/MyDrive/Atividade/products.csv')
brand = pd.read_csv('/content/drive/MyDrive/Atividade/brands.csv')
prod.head()
```

```python
preco = prod.groupby('brand_id')['list_price'].mean().reset_index()
preco_marcas = preco.merge(brand, on='brand_id', how='left')
preco_marcas
```

```python
bins = [preco_marcas['list_price'].min(), preco_marcas['list_price'].quantile(1/3), preco_marcas['list_price'].quantile(2/3), preco_marcas['list_price'].max()]
labels = ["Barato", "Médio", "Caro"]
preco_marcas['categoria_preco'] = pd.cut(preco_marcas['list_price'], bins=bins, labels=labels)
preco_marcas
```

```python
order = pd.read_csv('/content/drive/MyDrive/Atividade/order_items.csv')
order.head(100)
```

```python
produto = order.merge(prod, on='product_id', how='inner')
identificador = produto.merge(brand, on='brand_id', how='left')
quantidade_por_marca = identificador.groupby('brand_name')['quantity'].sum().reset_index()
print(quantidade_por_marca)
```
